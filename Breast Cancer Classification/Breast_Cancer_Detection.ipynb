# BREAST CANCER CLASSIFICATION USING DEEP LEARNING OR ARTIFICIAL NEURAL NETWORK

## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

## loading the dataset and storing it in a dataframe

```python
dataset = load_breast_cancer()
```

```python
dataset
```

The Overview of the dataset
- the dataset['data'] contains the data \
- the dataset['target'] contains the target \
- the dataset['target_names'] contains the target names \
- the dataset['feature_names'] contains the names of the feature(columns)

Target Representation \
- 0 - malignant (Cancer) \
- 1 - benign



```python
# converting the dataset into pandas dataframe
df = pd.DataFrame( dataset['data'], columns = dataset['feature_names'])
```

```python
df.head()
```

```python
# Adding the target column in the dataframe

df['Target'] = dataset['target']
```

## splitting the data into train and test data

```python
X = df.drop(columns=['Target'], axis = 1)
y = df['Target']
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
y_train.value_counts()
```

## BUILDING THE NEURAL NETWORK

```python
# importing the necessary libraries

import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
```

```python
# creating the model
model = Sequential()

# adding the layers to the model
model.add(Flatten(input_shape = (30,)))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(2,activation = 'sigmoid'))

# get the model summary (bias and weights)
model.summary()
```

```python
# Compiling OR Training the model

model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# TRAINING the model

history = model.fit(X_train, y_train, validation_split=0.1, epochs = 10)
```

**Accuracy = 73.83** without standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_test_std
```

```python
history_std = model.fit(X_train_std, y_train, validation_split=0.1, epochs = 10)
```

**Accuracy = 98.65** with *Standardization*

```python
# MAKING PREDICTION

y_pred = model.predict(X_test_std)
```

```python
y_pred[0]
```

model (sigmoid) provides with the propability with which it is sure \
- 99.10% -> malignant (Cancer)
- 82.165 -> bengin

```python
#predicting the labels for the output

y_pred_labels = [np.argmax(i) for i in y_pred]
```

```python
y_pred_labels[:15]
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred_labels, y_test)
```
